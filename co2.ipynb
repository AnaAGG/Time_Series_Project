```python
co2 = pd.read_csv("Data/co2.csv", sep = ";")
co2.head()
```

```python
columns = ['Data Source,World Development Indicators,a']
co2.columns = columns 
co2.head()
```

```python
co2.columns 
```

```python
foo = lambda x: pd.Series([i for i in reversed(x.split(','))])
rev = co2['Data Source,World Development Indicators,a'].apply(foo)
```

```python
rev = rev.replace('"','', regex=True)
```

```python
rev.columns = rev.iloc[1]
rev.drop(rev.index[1], inplace = True)
rev.head()
```

```python
rev.drop(["2015", "2016", "2017", "2018"], axis = 1, inplace = True)
```
